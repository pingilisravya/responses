```python
a = a.rename(columns = {"Name of the Organization":"col_3"})
  
a
```

```python
org=a.groupby("response_2")['response_3'].value_counts(normalize=False).unstack()
org.plot(kind='bar', alpha=1, stacked='False')

plt.title('response_2 VS response_3')
plt.ylabel('responses')
plt.show()
```

```python
orga=a.groupby("col_3")['response_2'].value_counts(normalize=False).unstack()
orga.plot(kind='bar', stacked='False',figsize=(10,5))

plt.title('name of organization')
plt.ylabel('no of responses')
plt.show()
```

```python
put=a.groupby("Name of the Respondent")['response_9'].value_counts(normalize=False).unstack()
put.plot(kind='bar', alpha=1, stacked='False')

plt.title('Name of the Respondent VS response_9')
plt.ylabel('research annually')
plt.show()
```

```python
sit=a.groupby("col_3")['9 Will you prefer outsourcing your projects to startup firm?'].value_counts(normalize=False).unstack()
sit.plot(kind='bar', alpha=1, stacked='False')

plt.title('Name of the org VS project outsourcing')
plt.ylabel('projects to start a firm')
plt.show()
```
